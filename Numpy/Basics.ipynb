```python
# numpy
import numpy as np

# Creating a NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Basic operations
arr = arr + 10

print(arr)

# Broadcasting and arithmetic operations
arr2 = np.array([10, 20, 30, 40, 50])
sum_arr = arr + arr2

print(sum_arr)

# Linear Algebra
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(a, b, sep='\n')

# Matrix multiplication
c = np.dot(a, b)

print(c)
```

```python
# Numpy and Matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Data for plotting
t = np.arange(0., 5., 0.2)
s = np.exp(-t) * np.cos(2 * np.pi * t)

print(t, s, sep='\n')

# Basic line plot
plt.plot(t, s)

# Customizing the plot
plt.title('Damped oscillation')
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Prepare the data (Can be adjusted depending on the type of plot)
x = np.linspace(0, 10, 100)  # For example, 100 points from 0 to 10
y = np.sin(x)                # Corresponding sine values
z = np.cos(x)

# 2. Create a figure and an axes (or subplots)
fig, ax = plt.subplots()

# 3. Plot the data
ax.plot(x, y, label='Sine wave')  # For a line plot
ax.plot(x, z, label='Cos wave')
# ax.scatter(x, y, label='Data points')  # For a scatter plot
# ax.bar(names, values, label='Bar chart')  # For a bar chart
# ... other types of plots

# 4. Customize the plot
ax.set_title('Title Here')           # Add a title
ax.set_xlabel('x label here')        # Add an x-label
ax.set_ylabel('y label here')        # Add a y-label
ax.legend()                          # Add a legend

# 5. Show the plot
plt.show()
```
